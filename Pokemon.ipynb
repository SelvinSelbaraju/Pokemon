```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('pokemon.csv',index_col = 'pokedex_number')
df.columns
```

```python
df['weight_kg'] = df['weight_kg'].fillna('beans')
```

```python
df[df['weight_kg'] == 'beans']
```

```python
df2 = pd.read_csv('pokemon(1).csv')
df2
```

```python
df2.columns
```

```python
df2cols = df2[['ndex','hp', 'attack', 'defense', 'spattack',
       'spdefense', 'speed', 'total', 'weight', 'height','forme']]
```

```python
df2cols['pokedex_number'] = df2cols['ndex']
df2cols.head()
```

```python
df3 = pd.merge(df,df2cols,on = 'pokedex_number')
```

```python
df3['weight'] = df3['weight'].apply(lambda x: float(x.split(' lbs.')[0])) 
```

```python
df3['height']
```

```python
df3['heightft'] = df3['height'].apply(lambda x: float(x.split('"')[0].split("'")[0]))
```

```python
df3['heightft']
```

```python
df3['heightinch'] = df3['height'].apply(lambda x: float(x.split('"')[0].split("'")[-1]))
df3['heightinch']
```

```python
df3
```

```python
df3['totalinches']=df3['heightft']*12+df3['heightinch']
```

```python
sns.distplot(df3['totalinches'])
```

```python
df3.columns
```

```python
df4 = df3.drop(['abilities', 'against_bug', 'against_dark', 'against_dragon',
       'against_electric', 'against_fairy', 'against_fight', 'against_fire',
       'against_flying', 'against_ghost', 'against_grass', 'against_ground',
       'against_ice', 'against_normal', 'against_poison', 'against_psychic',
       'against_rock', 'against_steel', 'against_water','classfication','experience_growth','height_m','japanese_name','weight_kg', 'generation','ndex', 'spattack', 'spdefense','height',
       'heightft', 'heightinch','hp_y', 'attack_y', 'defense_y', 'speed_y','total','percentage_male'],axis = 1)
```

```python
df4['type1'] = df4['type1'].apply(lambda x: x[0:3])
```

```python
plt.figure(figsize = (12,6))
sns.boxplot(data = df4, x = 'type1',y = 'attack_x')
plt.title('Boxplots showing the attack distribution for each type',fontdict = {'fontsize':20})
plt.xlabel('Type 1')
plt.ylabel('Attack')
```

```python
sns.barplot(data = df4,x = 'is_legendary',y = 'base_total',estimator = np.max)
```
